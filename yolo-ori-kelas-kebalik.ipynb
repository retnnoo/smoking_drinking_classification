```python
from collections import Counter
import glob
def class_counts(label_dir):
    cnt = Counter()
    for f in glob.glob(label_dir+'/*.txt'):
        for line in open(f):
            cls = int(line.split()[0]); cnt[cls]+=1
    return cnt
print("val class counts:", class_counts('/kaggle/input/trash-dataset-icra19/trash_ICRA19/dataset/val'))
print("val class counts dataset 2:", class_counts('/kaggle/input/underwater-trash-video-frame-extraction-datase/valid'))
print("test class counts:", class_counts('/kaggle/input/trash-dataset-icra19/trash_ICRA19/dataset/test'))
print("test class counts dataset 2:", class_counts('/kaggle/input/underwater-trash-video-frame-extraction-datase/test'))
```

```python
from ultralytics import YOLO
import os
import torch
import shutil
```

```python
from datetime import datetime

# === Konfigurasi ===
train_folder_name = "underwater_trash_detection" 
working_path = f"/kaggle/working/yolov8_training_runs/{train_folder_name}"
input_backup_path = f"/kaggle/input/yolo-backup/{train_folder_name}" 
backup_path = "/kaggle/working/yolo_backups"
os.makedirs(backup_path, exist_ok=True)

resume_mode = False

weights_working = f"{working_path}/weights/last.pt"
weights_input = f"{input_backup_path}/weights/last.pt"

if os.path.exists(weights_working):
    print("‚úÖ Checkpoint ditemukan di WORKING directory -> Resume training...")
    model = YOLO(weights_working)
    resume_mode = True
    
elif os.path.exists(weights_input):
    print("‚úÖ Checkpoint ditemukan di INPUT dataset -> Mengambil backup...")
    
    # Copy backup dari input ke working
    if os.path.exists(working_path):
        shutil.rmtree(working_path)
    shutil.copytree(input_backup_path, working_path)
    
    model = YOLO(f"{working_path}/weights/last.pt")
    resume_mode = True
    
else:
    print("üÜï Pelatihan baru dimulai...")
    model = YOLO("yolov8s.pt") 
    resume_mode = False

# === Auto Backup Callback ===
def custom_callback(trainer):
    current_epoch = trainer.epoch + 1
    if current_epoch % 5 == 0:

        for old_zip in glob.glob("/kaggle/working/yolo_backup_epoch_*.zip"):
            os.remove(old_zip)
            print(f"üóëÔ∏è Hapus ZIP lama: {old_zip}")
            
        if os.path.exists(backup_path):
            shutil.rmtree(backup_path)
        shutil.copytree(working_path, backup_path)
        zip_name = f"/kaggle/working/yolo_backup_epoch_{current_epoch}"
        shutil.make_archive(zip_name, 'zip', backup_path)

        print(f"‚úÖ Backup & ZIP selesai (epoch {current_epoch})!")
        print(f"üìÅ File ZIP: {zip_name}.zip siap didownload ‚úÖ")
```

```python
print(model.model)
```

```python
# data: Lokasi file data.yaml.
# epochs: Jumlah epoch (iterasi penuh melalui dataset pelatihan).
# imgsz: Ukuran gambar yang akan diresize saat input ke model. 640 adalah standar.
# batch: Jumlah gambar yang diproses per iterasi. Sesuaikan dengan memori GPU (16, 32, 64, dll.).
# name: Nama unik untuk folder hasil pelatihan di 'runs/detect/'.
# project: Nama folder induk untuk semua runs pelatihan  (default: 'runs').
# save_period: Frekuensi penyimpanan checkpoint model (misal, setiap 10 epoch).
# patience: Jumlah epoch untuk menunggu jika tidak ada peningkatan metrik validasi sebelum menghentikan pelatihan.
```

```python
print("Memulai pelatihan model...")
model.add_callback("on_train_epoch_end", custom_callback)
results = model.train(
    data='/kaggle/working/data.yaml',        
    epochs=100,              
    imgsz=640,               
    batch=16,                 
    name= train_folder_name, 
    project='yolov8_training_runs', 
    device=0,
    # Beberapa parameter opsional yang bisa ditambahkan:
    # workers=8,             # Jumlah worker untuk data loading (sesuaikan dengan CPU cores)
    patience=25,           # Menghentikan pelatihan jika tidak ada peningkatan selama 25 epoch
    # save_period=10,        # Simpan model setiap 10 epoch
    # cache=True,            # Cache gambar untuk pelatihan lebih cepat (memakan RAM)
    resume=resume_mode,           # Lanjutkan pelatihan dari checkpoint terakhir
)
print("Pelatihan selesai!")
```

```python
# Muat model terbaik yang sudah dilatih
best_model_path = os.path.join(results.save_dir, 'weights', 'best.pt')
trained_model = YOLO(best_model_path)
```

```python
# Evaluasi model pada dataset test
print("Mengevaluasi model pada data test...")
test_metrics = trained_model.val(split='test')
print(f"Metrics: {test_metrics}") # mAP50, mAP50-95, Precision, Recall.
```
